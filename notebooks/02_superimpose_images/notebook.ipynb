# 隣接するフレームを重ね合わせる
このノートではLoss関数を実装するために重要な画像の重ね合わせの方法を確認する。

何をやるのかを説明する前に、まず隣接する２フレームを見てみよう。
一見してこれらはあまり見た目上の違いがないように見える。

```python
from PIL import Image
image_t0 = Image.open("../../data/Scene01/15-deg-left/frames/rgb/Camera_0/rgb_00000.jpg")
image_t1 = Image.open("../../data/Scene01/15-deg-left/frames/rgb/Camera_0/rgb_00001.jpg")
```

```python
image_t0
```

```python
image_t1
```

しかし、実際は自車両が移動しているため、画素レベルでは大きな違いが生じている。
これを画素の差分で確認する。

```python
import numpy as np
import matplotlib.pyplot as plt

image_t0, image_t1 = np.array(image_t0).astype(np.int32), np.array(image_t1).astype(np.int32)
diff_t1_t0 = np.abs(image_t1 - image_t0)

plt.figure(figsize=(15, 15))
plt.imshow(diff_t1_t0)
print(np.sum(diff_t1_t0))
```

黒い箇所は画素の値の差が小さい場所だが、空と地面と壁は比較的に差分が少ないことがわかる。<br>
空は深度が大きく（つまり光線の反射元が遠い）自車両が多少進んでも変化がないためである。<br>
一方で地面と壁はテクスチャの変化が少ないためである。

上記の例で移動により画像が変化したことを確かめた。<br>
しかし、実は車両の姿勢・位置の変化および物体の表の３次元座標がわかっているならば、変化をキャンセルして画像を重ね合わせることが可能である。<br>

ここではDepth、カメラパラメタをつかい、この重合わせを実行してみる。<br>
この操作は深度推定の学習時に`photometric loss`という誤差を計算するために使われるので重要である。

この操作を理解するためにはカメラの投影モデルを理解している必要がある。<br>
カメラの投影モデルは以下の式のように、カメラ座標（３次元）から画像平面（２次元）への投影として内部パラメタ・外部パラメタの行列をかけることで表される。

![projection](http://opencv.jp/opencv-2.1/cpp/_images/math/a7fb9389364911f79ec6fb5067b4b49d60b9940c.png)  

画像平面（2D）からカメラ座標（3D）への投影はその逆だが、失われた次元の復元を（３次復元と呼ぶ）するときに深度が必要になる。

今回は隣接する時刻t+0とt+1のフレームを利用する。t+1のフレームをt+0重ね合わせてみる。
重ね合わせには`cv2.remap()`を用いる。remapは一見してわかりにくい関数だが、これから説明する使用方法をみれば直感的に理解できるようになるはずである。

t+0とt+1の外部パラメタの読み込みを行う。
ちなみに、内部パラメタはフレーム間では変化しない点に注意されたい。

```python
import pandas as pd

# 外部パラメタを読み込む
data = pd.read_csv("../../data/vkitti_2.0.3_textgt/Scene01/15-deg-left/extrinsic.txt", delimiter=' ')

row_t0 = data.loc[0]  # frame #0, cameraID #0
row_t1 = data.loc[2]  # frame #1, cameraID #0

# t+0の外部パラメタ
extrinsic_t0 = np.array([
    [row_t0["r1,1"], row_t0["r1,2"], row_t0["r1,3"], row_t0["t1"]],
    [row_t0["r2,1"], row_t0["r2,2"], row_t0["r2,3"], row_t0["t2"]],
    [row_t0["r3,1"], row_t0["r3,2"], row_t0["r3,3"], row_t0["t3"]],
    [row_t0["0"],    row_t0["0.1"],  row_t0["0.2"],  row_t0["1"]]])

# t+1の外部パラメタ
extrinsic_t1 = np.array([
    [row_t1["r1,1"], row_t1["r1,2"], row_t1["r1,3"], row_t1["t1"]],
    [row_t1["r2,1"], row_t1["r2,2"], row_t1["r2,3"], row_t1["t2"]],
    [row_t1["r3,1"], row_t1["r3,2"], row_t1["r3,3"], row_t1["t3"]],
    [row_t1["0"],    row_t1["0.1"],  row_t1["0.2"],  row_t1["1"]]])
```

```python
extrinsic_t0, extrinsic_t1
```

```python
data = pd.read_csv("../../data/vkitti_2.0.3_textgt/Scene01/15-deg-left/intrinsic.txt", delimiter=' ')
row0 = data.loc[0]

intrinsic = np.array([
    [row0["K[0,0]"],            0.0, row0["K[0,2]"]],
    [           0.0, row0["K[1,1]"], row0["K[1,2]"]],
    [           0.0,            0.0,           1.0]])
intrinsic
```

## t+0の画像平面からワールド座標への投影
remap用のmapを作るために、t+0における画像座標をまずはワールド座標に投影する。

画像座標からワールド座標への投影は先程のカメラの投影モデルの逆操作にあたるので、内部パラメタ行列と外部パラメタ行列それぞれの逆行列をもとめておく。

```python
intrinsic_inv = np.linalg.inv(intrinsic)
extrinsic_t0_inv = np.linalg.inv(extrinsic_t0)
intrinsic_inv, extrinsic_t0_inv
```

３次元復元用にdepthをnumpyの行列として読みこむ。<br>
また、外部パラメタにあわせてメータ単位にするために深度を100で割っておく。

```python
depth = Image.open("../../data/Scene01/15-deg-left/frames/depth/Camera_0/depth_00000.png")
depth = np.array(depth) / 100
depth
```

まずは画像の画素ごとにその座標が入った行列を作成する。<br>
t+0における各画素がt+1の画像のどの座標の画素と同じなのかをという関係を表すmapをcv2.remap()に与えることで画像の重ね合わせがを行うが、その前準備である。

```python
xi = np.arange(0, depth.shape[1], 1)
yi = np.arange(0, depth.shape[0], 1)
coord_x, coord_y = np.meshgrid(xi, yi)
coord_z = np.ones_like(coord_x)
image_coords_t0 = np.stack([coord_x, coord_y, coord_z], axis=-1)
image_coords_t0 = image_coords_t0.astype(np.float64)

image_coords_t0.shape
```

画素の座標を内部パラメタの逆行列をかけることでカメラ座標に投影する。

```python
camera_coords_t0 = intrinsic_inv @ image_coords_t0.transpose(0, 2, 1)
camera_coords_t0 = camera_coords_t0.transpose(0, 2, 1)
```

３次元復元を行う。

```python
camera_coords_t0  *= depth[..., np.newaxis]
```

```python
ones = np.ones((camera_coords_t0.shape[0], camera_coords_t0.shape[1], 1))
camera_coords_t0 = np.concatenate([camera_coords_t0, ones], axis=-1)
camera_coords_t0.shape
```

外部パラメタの逆行列をかけてカメラ座標からワールド座標へ投影する

```python
world_coords = extrinsic_t0_inv @ camera_coords_t0.transpose(0, 2, 1)
world_coords = world_coords.transpose(0, 2, 1)
world_coords.shape
```

## ワールド座標からt+1の画像平面へ投影
ここまででt+0の画像平面上の座標がワールド座標に投影した。<br>
次からワールド座標をt+1の画像座標へ投影する。

外部パラメタ行列をワールド座標にかけてカメラ座標に投影する。

```python
camera_coords_t1 = extrinsic_t1[:3, :] @ world_coords.transpose(0, 2, 1)
camera_coords_t1 = camera_coords_t1.transpose(0, 2, 1)
camera_coords_t1.shape
```

カメラ座標において３次元復元の逆の操作をおこない、同次座標に変換する。<br>
これはZ軸がすべて1.0になるようにすればよい。

```python
camera_coords_t1 /= camera_coords_t1[..., [2]]
```

内部パラメタ行列をかけて画像座標に投影する。

```python
image_coords_t1 = intrinsic @ camera_coords_t1.transpose(0, 2, 1)
image_coords_t1 = image_coords_t1.transpose(0, 2, 1)
image_coords_t1.shape
```

これまでの処理によりt+0の各画素に対してt+1の画素の対応関係を表すmapを生成した。
このmapを`cv2.remap()`に入力し、t+0の画像とt+1の画像を重ね合わせる。

```python
import cv2
map_x_t1 = image_coords_t1[..., 0].astype(np.float32)
map_y_t1 = image_coords_t1[..., 1].astype(np.float32)

image_t1_to_t0 = cv2.remap(image_t1.astype(np.uint16), map_x_t1, map_y_t1, interpolation=cv2.INTER_LINEAR)
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(image_t1_to_t0)
```

t+1の画像をt+0に重ね合わせたので、t+1の視点では視界から外れた部分は対応する画素が存在せず、画像の下端などが黒く塗りつぶされていることがわかる。

## 重ね合わせの差分
t+0, t+1の画像は何も変更せずに差分を見ると、様々な箇所で差がでていたが、今回重ね合わせることで差が小さくなったはずである。<br>
それを確かめる。

```python
diff_t0t1_t0 = np.abs(image_t1_to_t0 - image_t0)
fig = plt.figure(figsize=(15, 10))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
ax1.title.set_text(f"simple diff abs_diff={np.sum(diff_t1_t0)}")
ax1.imshow(diff_t1_t0)
ax2.title.set_text(f"fsuper imposed abs_diff={np.sum(diff_t0t1_t0)}")
ax2.imshow(diff_t0t1_t0)
print()
```

今回の差分（下）は（視野から外れた部分以外は）フレーム間の単純な差分に比べて黒くなっていることから重ね合わせがうまくいっていることがわかる。<br>

注意深く見てみると、今回の差分においても大きな差が出ている箇所がいくつかある。<br>
例えば木の後ろ、右側の建物の出っ張りであるが、これは物陰などで片方の画像には存在しないものによって生じている。<br>
窓ガラスも差分が大きい。
これらは原理的に重ね合わせることができない。
